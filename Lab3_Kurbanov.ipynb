````python
2. Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры
````

````python
import numpy as np
import matplotlib.pyplot as plt

def parabolic(x):
    return 2 * (x ** 2) + 3 * x + 1

x_values = np.linspace(-20, 20, 400)

y_values = parabolic(x_values)

plt.figure(figsize=(10, 7))
plt.plot(x_values, y_values, color='blue', linewidth=2, linestyle='-', label='y = 2x² + 3x + 1')

plt.title('График параболы', fontsize=18)
plt.xlabel('Ось X', fontsize=14)
plt.ylabel('Ось Y', fontsize=14)

plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper right')

plt.show()
````

````python
3. Построить график произвольной гармонической функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры
````

````python
import numpy as np
import matplotlib.pyplot as plt

def sine_wave(t):
    return np.sin(2 * np.pi * t)

t_values = np.linspace(0, 4, 400)

wave_values = sine_wave(t_values)

plt.figure(figsize=(10, 7))
plt.plot(t_values, wave_values, color='orange', linewidth=2, linestyle='-', label='y = sin(2πt)')

plt.title('График синусоиды', fontsize=18)
plt.xlabel('Параметр t', fontsize=14)
plt.ylabel('Значение синуса', fontsize=14)

plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper right')

plt.show()
````

````python
4.	Вывести графики п.2 и 3 в разные поля при помощи функции subplot()
````

````python
import numpy as np
import matplotlib.pyplot as plt

def parabola(x):
    return 2 * x**2 + 3 * x + 1

def sine_wave(t):
    return np.sin(2 * np.pi * t)

figure, axes = plt.subplots(2, 1, figsize=(10, 12))

x_parabola = np.linspace(-10, 10, 400)
y_parabola = parabola(x_parabola)

axes[0].plot(x_parabola, y_parabola, color='green', linewidth=2, linestyle='-', label='y = 2x² + 3x + 1')
axes[0].set_title('График параболы', fontsize=18)
axes[0].set_xlabel('Ось X', fontsize=14)
axes[0].set_ylabel('Ось Y', fontsize=14)
axes[0].grid(True, linestyle='--', alpha=0.7)
axes[0].legend()

x_sine = np.linspace(0, 4, 400)
y_sine = sine_wave(x_sine)

axes[1].plot(x_sine, y_sine, color='blue', linewidth=2, linestyle='-', label='y = sin(2πt)')
axes[1].set_title('График синусоиды', fontsize=18)
axes[1].set_xlabel('Параметр t', fontsize=14)
axes[1].set_ylabel('Значение синуса', fontsize=14)
axes[1].grid(True, linestyle='--', alpha=0.7)
axes[1].legend()

plt.tight_layout()
plt.show()
````

````python
5.	Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.
````

````python
x_values_quadratic = np.linspace(-10, 10, 400)
y_values_quadratic = parabola(x_values_quadratic)

x_values_harmonic = np.linspace(0, 4, 400)
y_values_harmonic = sine_wave(x_values_harmonic)

plt.figure(figsize=(10, 7))

plt.plot(x_values_quadratic, y_values_quadratic, label='y = 2x² + 3x + 1', color='red', linewidth=2)

plt.plot(x_values_harmonic, y_values_harmonic, label='y = sin(2πt)', linestyle='--', color='blue', linewidth=2)

plt.title('Графики параболической и гармонической функций', fontsize=18)
plt.xlabel('Значения X', fontsize=14)
plt.ylabel('Значения Y', fontsize=14)

plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
````

````python
6.
x(t) = cos(3*t)` и `y=sin(2*t)`, `t` принадлежит `[0 .. 2*pi]
````

````python

t = np.linspace(0, 2 * np.pi, 400)  
x_parametric = np.cos(3 * t) 
y_parametric = np.sin(2 * t) 


plt.figure(figsize=(8, 6))
plt.plot(x_parametric, y_parametric)

plt.title('x(t) = cos(3*t), y(t) = sin(2*t)', fontsize=16)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)

plt.grid(True)

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

x_range = np.linspace(-5, 5, 400)
y_range = np.linspace(-5, 5, 400)
x_mesh, y_mesh = np.meshgrid(x_range, y_range)

z_mesh = (x_mesh ** 2 - y_mesh ** 2)

figure = plt.figure(figsize=(10, 8))
axis = figure.add_subplot(111, projection='3d')

surface = axis.plot_surface(x_mesh, y_mesh, z_mesh, cmap='viridis')

axis.set_title('Гиперболический параболоид', fontsize=18)
axis.set_xlabel('Ось X', fontsize=14)
axis.set_ylabel('Ось Y', fontsize=14)
axis.set_zlabel('Ось Z', fontsize=14)

plt.show()
````
